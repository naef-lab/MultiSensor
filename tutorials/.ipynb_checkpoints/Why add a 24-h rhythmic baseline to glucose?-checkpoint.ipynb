# Why add a 24-h rhythmic baseline to glucose?

While presenting the project in different environments, I've noticed that the underlying 24-hour rhythm in glucose levels (in Model 1) sometimes causes confusion. The aim of this short tutorial is just to illustrate the model building process that lead to the inclusion of this term.

Our first attempt at modelling the data included only the contribution of meals. As an illustration, there is a class called Model1SingleSensor_nocirc that represents Model 1 without the 24-hour rhythmic term.

Let's load the data and plot the model fit, in a similar manner to the main tutorial but using the class without the 24-hour rhythm.

```python
import os
import dill as pickle
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import sys
import tensorflow_probability as tfp
import MSS.config as config
from tensorflow_probability.python import bijectors as tfb
from scipy.stats.stats import pearsonr
from tensorflow_probability import distributions as tfd
from MSS.model1dataset import DatasetModel1SingleSensor
from MSS.model1 import Model1SingleSensor
from MSS.model1 import Model1SingleSensor_nocirc
from scipy.optimize import curve_fit
os.environ['TF_XLA_FLAGS'] = '--tf_xla_enable_xla_devices'
```

```python
fname_gluc1 = 'tutorial-data/glucose_MSS02-1.csv'
fname_food1 = 'tutorial-data/food_MSS02-1.xlsx'
data_gluc = DatasetModel1SingleSensor(fname_gluc1,
                                      fname_food1,
                                      study_id='02',
                                      sensor='1')
```

## Data overview

```python
data_gluc.plot_glucose_overview()
```

## Fitting the model without the circadian component

To fit the model without the underlying circadian component, here we will use a class Model1SingleSensor_nocirc

```python
model1 = Model1SingleSensor_nocirc(data_gluc)
model1.find_MAP(model1.get_loss_and_grads_map,N_starts=200,print_on=False)
```

## Checking the model fit

We can now compare the fitted model that only uses meals to the data. The Pearson correlation coefficient between the data and model fit is displayed in the figure title as R. The data is shown in blue, and the model using only the meals (i.e. without the underlying circadian trend) is shown in orange.

```python
model1.check_meals_glucose_fit(model1.params_opted_map)
```

We can also compare the model and data by visualising the residual.

```python
def func(t, A0, A1, phix):
    """Cosinor circadian function"""
    w = 2*np.pi/24
    return A0 + A1 * (np.cos(w * t - phix ) +1)/2

def plot_residual(model1):
    """Calculates and plots the residual of the model (i.e. data - model prediction)"""
    
    params = model1.params_opted_map
    f_meals_and_circadian, f_meals, f_circadian = model1.all_mean_funcs(params,model1.X1,
                                                                          model1.T_sub_timestamp1,
                                                                          model1.food_item_index1,1)
    X = model1.X1.numpy()/24
    Y = model1.Y1.numpy()[:,0]

    residual = Y-f_meals_and_circadian

    plt.plot(X,residual,linewidth = 1,label='Residual')

    initial_guess = [-1, 2, -3]  # Initial guess for [A0, A1, phix, w]
    popt, pcov = curve_fit(func, X*24, residual.numpy(), p0=initial_guess)

    A0_fit, A1_fit, phix_fit = popt

    t_test = X*24
    y_test = func(t_test, A0_fit, A1_fit, phix_fit)

    plt.plot(t_test/24,y_test,'k',label='Circadian fit')


    x_min, x_max = plt.xlim()
    plt.xticks(range(int(x_min), int(x_max) + 1))

    plt.title('Residual (data - model fit)')
    plt.xlabel('Time (days)')
    plt.ylabel('Glucose (mmol/L)')
    plt.ylim([-2,2])
    
plot_residual(model1)    
```

As we can see, the residual seems to show a 24-hour structure, and the black line shows a 24-hour cosinor fit to the residual. This indicates a systematic error in the model. To better capture this behaviour, our solution was simply to add an underlying 24-hour rhythm in the data, which is described as Model 1 in the manuscript.

## Recalculating the residual using Model 1 (with underlying 24-hour baseline glucose rhythm)

Now let's do the same thing but using the full Model 1 (i.e. with the underlying 24-hour baseline glucose rhythm included).

```python
model2 = Model1SingleSensor(data_gluc)
model2.find_MAP(model2.get_loss_and_grads_map,N_starts=200,print_on=False)
```

```python
model2.check_meals_glucose_fit(model1.params_opted_map)
```

The correlation between the model fit and the data has increased from 0.68 to 0.77 using the full model, and we can see that the model fit has improved visibly.

Now let's check the model residual.

```python
plot_residual(model2)
```

We can see that the 24-hour structure in the residual is decreased, which was the motivation for including the 24-hour rhythm in the model.
